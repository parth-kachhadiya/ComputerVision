```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
base_img = cv2.imread('Resources/DATA/dog_backpack.png')
img = cv2.cvtColor(base_img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
def draw_circle(event, x, y, flags, param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img, (x, y), 100, (0, 0 ,255), 5)
```

```python
cv2.namedWindow(winname = "DrawUponDog")
cv2.setMouseCallback('DrawUponDog', draw_circle)
```

```python
while True:
    cv2.imshow('DrawUponDog', img)
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```

#### ` : Image is not in proper color order. Because 'matplotlib' & 'opencv' expects different order of RGB`

```python
# Converting the channel of the image from 'BGR' to 'RGB'
fixed_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## **Manipulating size of image**

```python
new_img = cv2.resize(fixed_img, (1000, 400))
```

```python
plt.imshow(new_img)
```

```python
# Manipulate size via ratio.
width_ratio = 0.8 # 80%
height_ratio = 0.2 # 20%
new_img = cv2.resize(fixed_img, (0, 0), fixed_img, width_ratio, height_ratio)
```

```python
plt.imshow(new_img)
new_img.shape
```

## **Flipping image**

```python
new_img = cv2.flip(fixed_img, 0) # on 'y' axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fixed_img, 1) # on 'x' axis
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fixed_img, -1) # on both side..
plt.imshow(new_img)
```

## **Drawing on image**

```python
blank_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1 = (200, 400), pt2 = (400, 200), color = (0,255, 0), thickness = 10) # Changes will be happen in original image
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1 = (200, 200), pt2 = (300, 300), color = (0,0,255), thickness = 5)
plt.imshow(blank_img)
```

```python
cv2.circle(img = blank_img, center = (100, 100), radius = 50, color = (255, 0, 0), thickness = 7)
plt.imshow(blank_img)
```

```python
cv2.circle(img = blank_img, center = (400, 400), radius = 50, color = (255, 0, 0), thickness = -1)
plt.imshow(blank_img)
```

```python
cv2.line(img = blank_img, pt1 = (0, 0), pt2 = (500, 500), color = (102, 255, 255), thickness = 5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text = "Parth", org = (10, 500), fontFace = font, fontScale = 4, color = (255,255,255), thickness = 2, lineType = cv2.LINE_AA)
plt.imshow(blank_img)
```

## **Custom polygons**

```python
blank_img = np.zeros(shape = (512, 512, 3), dtype = np.int16)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype = np.int32)
```

```python
vertices, vertices.shape
```

```python
pts = vertices.reshape((-1, 1, 2))
```

```python
pts, pts.shape
```

```python
cv2.polylines(blank_img, [pts], isClosed = True, color = (255, 0, 0), thickness = 5)
plt.imshow(blank_img)
```

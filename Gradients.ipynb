# **Gradients**

- **An image gradient is a directional change in the intensity or color in an image**
- **Gradients is the extention of morphological operator**
- **Understanding gradients leads us to understanding `edge detection`, `object detection`, `object tracking`, `classification`**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img ,cmap='gray')
```

```python
img = cv2.imread("Resources/DATA/sudoku.jpg", 0)
display_img(img)
```

```python
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5) # 'X' gradient (vertical lines)
```

```python
display_img(sobelx)
```

```python
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5) # 'Y' gradient (horizontal lines)
```

```python
display_img(sobely)
```

```python
lplacian = cv2.Laplacian(img, cv2.CV_64F) # "Edge detection" of 'X' & 'Y'
```

```python
display_img(lplacian)
```

```python
blended = cv2.addWeighted(src1=sobelx, alpha=0.5, src2=sobely, beta=0.5, gamma=0)
```

```python
display_img(blended)
```

```python
ret, th1 = cv2.threshold(img, 130, 255, cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4, 4), np.uint8)
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```

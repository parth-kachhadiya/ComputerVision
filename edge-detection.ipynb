## **Canny edge detection**
- **Developed in 1968 by `John Canny` & is a multi-stage algorithm.**
- **One of the most popular edge detection algorithms.**

### **Process**
- **Apply gaussian filter to smooth the image in order to remove the noise.**
- **Apply non-maximum suppression to get rid of spurious response to edge detection**
- **Apply double threshold to determine potential edges**
- **Track edge by hysteresis: FInalize the detection of edges by suppressing all the other edges that are weak and not connected to strong edges.**

**This algorithm also requires a user to decide on low & high threshold vlaues.**

**`NOTE`** : *For high resolution images where you only want general edges, it is usually a good idea to apply a `custom blur` before applying the Canny Algorithm.*

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("../Image_processing/Resources/DATA/sammy_face.jpg")
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img, threshold1=127, threshold2=127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=255)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=0, threshold2=127)
plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
# Lower threshold to either 0 or 70% of the median value whichever is grater
lower = int(max(0, 0.7 * med_val))

# Upper threshold to either 130% of the median or the max 255, whichever is smaller
upper = int(min(255, 1.3 * med_val))
```

```python
lower, upper
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img, threshold1=lower, threshold2=upper + 100)
plt.imshow(edges)
```

```python
blured_img = cv2.blur(img, ksize=(5, 5))
```

```python
edges = cv2.Canny(image=blured_img, threshold1=lower, threshold2=upper + 50)
plt.imshow(edges)
```
